```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

#to print all the length of the RGB channels of the images
np.set_printoptions(threshold=np.inf)
```

```python
#First exercise
# load images
images = []
for element in range(5):
    filename = "test_images/" + str(element) + ".png"
    images.append(cv2.cvtColor(cv2.imread(filename), cv2.COLOR_BGR2RGB))

#show images
for image in images:
    plt.figure()
    plt.imshow(image)
```

```python
#Get RGB channels
r_channel_images = []
g_channel_images = []
b_channel_images = []

for image in images:
    tmp_r = np.empty(0)
    tmp_g = np.empty(0)
    tmp_b = np.empty(0)
    for pixel in image:
        tmp_r = np.append(tmp_r,pixel[:,0])
        tmp_g = np.append(tmp_g,pixel[:,1])
        tmp_b = np.append(tmp_b,pixel[:,2])
    r_channel_images.append(tmp_r)
    g_channel_images.append(tmp_g)
    b_channel_images.append(tmp_b)


print("RGB channels of the images:")
for idx in range(5):
    print(str(idx)+". image R channel:")
    print(r_channel_images[idx])
    print(str(idx)+". image G channel:")
    print(g_channel_images[idx])
    print(str(idx)+". image B channel:")
    print(b_channel_images[idx])
    
```

```python
#Second exercise
#Calculating the mean, and variance of the given channels.
r_channel_means = []
r_channel_vars = []
g_channel_means = []
g_channel_vars = []
b_channel_means = []
b_channel_vars = []

for img in r_channel_images:
    r_channel_means.append(img.mean())
    r_channel_vars.append(img.var())

for img in g_channel_images:
    g_channel_means.append(img.mean())
    g_channel_vars.append(img.var())
    
for img in b_channel_images:
    b_channel_means.append(img.mean())
    b_channel_vars.append(img.var())

#Transforming the data
standard_r = []
standard_g = []
standard_b = []

for idx in range(5):
    r = r_channel_images[idx]
    normed_r = (r - r.mean()) / np.std(r)
    normed_r -= normed_r.mean()
    standard_r.append(normed_r)
    
    g = g_channel_images[idx]
    normed_g = (g - g.mean()) / np.std(g)
    normed_g -= normed_g.mean()
    standard_g.append(normed_g)
    
    b = b_channel_images[idx]
    normed_b = (b - b.mean()) / np.std(b)
    normed_b -= normed_b.mean()
    standard_b.append(normed_b)

#Checking the data

print("Mean:")
for idx in range(5):
    print(standard_r[idx].mean())
    print(standard_g[idx].mean())
    print(standard_b[idx].mean())
#There are some numbers that are not 0.0, but the are really small, all in the 10^(-17) scale.
#Maybe they are ~= 0 ? Some number bugs?
print("")
print("Variance:")
for idx in range(5):
    print(standard_r[idx].var())
    print(standard_g[idx].var())
    print(standard_b[idx].var())
#Same thing, there are some numbers that are not 1.0, but are rather close to it?
    
```

```python
#Third exercise
#Read, and print the written content of the webpage
import bs4
import urllib.request

link = "https://blog.keras.io/the-future-of-deep-learning.html"
webpage=str(urllib.request.urlopen(link).read())
soup = bs4.BeautifulSoup(webpage)

text = str(soup.get_text()).replace("\\n","")
text = text.split("The future of deep learning")[2]
text = text.split("var gaJsHost")[0]

text = text.lower()
import collections
frequency_per_char = collections.Counter(text)
d = {key: value for key, value in dict(frequency_per_char).items() if key.isalpha()}
import matplotlib.pyplot as plt

keys = d.keys()
vals = d.values()

plt.bar(keys, np.divide(list(vals), sum(vals)), label="Real distribution")
plt.show()
```

```python
#Fourth exercise
from scipy import signal
from scipy.io import wavfile

sample_rate, samples = wavfile.read('sample_sound.wav')
frequencies, times, spectrogram = signal.spectrogram(samples, sample_rate)

plt.pcolormesh(times, frequencies, spectrogram)
plt.imshow(spectrogram)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
